```python
import pandas as pd
import numpy as np
# Data Mentah dari file log
log_data = """
PLAYER_JOIN;ID_Pemain=101;Nama_Pemain=Adi;Tanggal_Bergabung=2025-08-20
PLAYER_JOIN;ID_Pemain=102;Nama_Pemain=Budi;Tanggal_Bergabung=2025-08-21
CHAR_CREATE;ID_Karakter=201;ID_Pemain=101;Kelas=Warrior;Level=50
ACTIVITY;ID_Aktivitas=1;ID_Karakter=201;Tipe_Aktivitas=Dungeon Clear;Gold_Didapat=1500
PLAYER_JOIN;ID_Pemain=103;Nama_Pemain=Cici;Tanggal_Bergabung=2025-08-22
CHAR_CREATE;ID_Karakter=202;ID_Pemain=102;Kelas=Mage;Level=55
ERROR;Invalid log entry;CODE=501
CHAR_CREATE;ID_Karakter=203;ID_Pemain=101;Kelas=Rogue;Level=45
ACTIVITY;ID_Aktivitas=2;ID_Karakter=202;Tipe_Aktivitas=PVP Win;Gold_Didapat=750

ACTIVITY;ID_Aktivitas=3;ID_Karakter=203;Tipe_Aktivitas=Dungeon Clear;Gold_Didapat=1200
PLAYER_JOIN;ID_Pemain=104;Nama_Pemain=Dedi
"""
```

```python
def parse_game_log(data):
    list_pemain = []
    list_char = []
    list_aktivitas = []
    list_baris = data.strip().split('\n')
    for baris in list_baris:
        if baris.startswith('PLAYER_JOIN'):
            bagian_data_pemain = baris.split(';')[1:]
            pemain_baru = {key: val for key, val in (p.split('=') for p in bagian_data_pemain)}
            list_pemain.append(pemain_baru)
        elif baris.startswith('CHAR_CREATE'):
            bagian_data_char = baris.split(';')[1:]
            char_baru = {key: val for key, val in (c.split('=') for c in bagian_data_char)}
            list_char.append(char_baru)
        elif baris.startswith('ACTIVITY'):
            bagian_data_act = baris.split(';')[1:]
            act_baru = {key: val for key, val in (a.split('=') for a in bagian_data_act)}
            list_aktivitas.append(act_baru)
    return list_pemain, list_char, list_aktivitas
list_pemain, list_char, list_aktivitas = parse_game_log(log_data)
print(parse_game_log(log_data))
```

```python
df_pemain = pd.DataFrame(list_pemain)
df_char = pd.DataFrame(list_char)
df_activity = pd.DataFrame(list_aktivitas)

display(df_pemain.head())
df_pemain.info()
display(df_pemain.describe())
display(df_char.head())
df_char.info()
display(df_char.describe())
display(df_activity.head())
df_activity.info()
display(df_activity.describe())
```

```python
df_pemain_bersih = df_pemain.copy().dropna(subset=['Tanggal_Bergabung'])
df_pemain_bersih['ID_Pemain'] = df_pemain_bersih['ID_Pemain'].astype(int)
df_char_bersih = df_char.copy()
list_untuk_char = ['ID_Karakter', 'ID_Pemain', 'Level']
df_char_bersih[list_untuk_char] = df_char_bersih[list_untuk_char].astype(int)
df_act_bersih = df_activity.copy()
list_untuk_act = ['ID_Aktivitas', 'ID_Karakter', 'Gold_Didapat']
df_act_bersih[list_untuk_act] = df_act_bersih[list_untuk_act].astype(int)
df_pemain_bersih.info()
df_char_bersih.info()
df_act_bersih.info()
```

```python
df_gabungan_sementara = pd.merge(df_char_bersih, df_act_bersih, on='ID_Karakter', how='inner')
df_master = pd.merge(df_pemain_bersih, df_gabungan_sementara, on='ID_Pemain', how='inner')
df_master
```

```python
laporan_akhir = df_master.groupby('Nama_Pemain').agg(
    Total_Gold=('Gold_Didapat', 'sum'),
    Rata_Rata_Level=('Level', 'mean')
)
laporan_akhir_terurut = laporan_akhir.sort_values(by=['Total_Gold', 'Rata_Rata_Level'], ascending=False)
print("--- 3 Pemain teratas berdasarkan total perolehan gold dan rata-rata level karakter adalah ---")
display(laporan_akhir_terurut.head())
```
