```python
import pandas as pd
import numpy as np

data_karakter = {
    'Nama_Karakter': ['Arion', 'Lira', 'Grom', 'Elara', 'Zane', 'Faye'],
    'Kelas': ['Warrior', 'Mage', 'Warrior', 'Mage', 'Rogue', 'Rogue'],
    'Level': [85, 88, 82, 90, 86, 87],
    'Attack_Power': [1200, 1550, 1100, np.nan, 1350, 1400],
    'Item_Value': [25000, 32000, 23000, 35000, 28000, np.nan]
}

df_karakter = pd.DataFrame(data_karakter)
```

```python
display(df_karakter.head())
display(df_karakter.info())
display(df_karakter.describe())
```

```python
df_bersih = df_karakter.copy()
mean_attack = df_bersih['Attack_Power'].mean()
mean_item_value = df_bersih['Item_Value'].mean()
df_bersih['Attack_Power'] = df_bersih['Attack_Power'].fillna(mean_attack)
df_bersih['Item_Value'] = df_bersih['Item_Value'].fillna(mean_item_value)
print(f"--- Info DataFrame Setelah Pembersihan ---")
df_bersih.info()
print("\n--- Tampilan DataFrame Bersih ---")
display(df_bersih)
```

```python
laporan_per_kelas = df_bersih.groupby('Kelas').agg(
    Total_Item_Value=('Item_Value', 'sum'),
    Rerata_Attack_Power=('Attack_Power', 'mean')
)
laporan_final = laporan_per_kelas.sort_values(by='Total_Item_Value', ascending=False)
print("--- Laporan Final Kekuatan per Kelas ---")
display(laporan_final)
```
