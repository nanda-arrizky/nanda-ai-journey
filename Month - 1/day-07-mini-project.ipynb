```python
import random

pilihan_nama_depan = ["Aria", "Zane", "Elara", "Kael"]
pilihan_nama_belakang = ["Shadowwind", "Ironheart", "Stormcaller", "Lightbringer"]
pilihan_kelas = ["Warrior", "Mage", "Rogue", "Healer"]
pilihan_senjata = {
    "Warrior": ["Greatsword", "Battle Axe"],
    "Mage": ["Fire Staff", "Ice Wand"],
    "Rogue": ["Twin Daggers", "Shadow Bow"],
    "Healer": ["Sacred Tome", "Holy Scepter"]
}

def buat_karakter():
    nama_depan = random.choice(pilihan_nama_depan)
    nama_belakang = random.choice(pilihan_nama_belakang)
    kelas_karakter = random.choice(pilihan_kelas)

    nama_lengkap = nama_depan + " " + nama_belakang
    senjata_tersedia = pilihan_senjata[kelas_karakter]
    senjata_pilihan = random.choice(senjata_tersedia)

    total_poin = 100

    kekuatan = random.randint(1, 80)
    sisa_poin = total_poin - kekuatan
    sihir = random.randint(1, sisa_poin)
    kelincahan = sisa_poin - sihir #Logicnya, karena angka sisanya sudah pasti dan ini adalah stats terakhir, maka tidak perlu pakai random.randint

    if kelas_karakter == "Warrior":
        kekuatan += 10
    elif kelas_karakter == "Mage":
        sihir += 10
    elif kelas_karakter == "Rogue":
        kelincahan += 10
    elif kelas_karakter == "Healer":
        sihir += 5
        kekuatan += 5
    
    print("Nama:", nama_lengkap)
    print("Kelas:", kelas_karakter)
    print("Senjata:", senjata_pilihan)
    print(f"Statistik -> Kekuatan: {kekuatan}, Sihir: {sihir}, Kelincahan: {kelincahan}")

print("--- Menciptakan Karakter Uji Coba #1 ---")
buat_karakter()

print("\n--- Menciptakan Karakter Uji Coba #2 ---")
buat_karakter()
```

```python
import random

# --- Pilihan yang Diperluas ---

# 10 Nama Depan Baru
pilihan_nama_depan_2 = ["Aria", "Zane", "Elara", "Kael", "Lyra", "Ronan", "Seraphina", "Jax", "Thorne", "Astrid"]

# 10 Nama Belakang Baru
pilihan_nama_belakang_2 = [
    "Shadowwind", "Ironheart", "Stormcaller", "Lightbringer", "Nightshade", 
    "Flamecrest", "Winterfell", "Voidgazer", "Sunstrider", "Stonehand"
]

# 3 Kelas Baru Ditambahkan
pilihan_kelas_2 = ["Warrior", "Mage", "Rogue", "Healer", "Paladin", "Hunter", "Warlock"]

# 1 Senjata Baru untuk Setiap Kelas (Termasuk Kelas Baru)
pilihan_senjata_2 = {
    "Warrior": ["Greatsword", "Battle Axe", "Warhammer"],
    "Mage": ["Fire Staff", "Ice Wand", "Arcane Orb"],
    "Rogue": ["Twin Daggers", "Shadow Bow", "Poisoned Crossbow"],
    "Healer": ["Sacred Tome", "Holy Scepter", "Blessed Mace"],
    "Paladin": ["Radiant Lance", "Shield of Dawn", "Lightforged Hammer"],
    "Hunter": ["Longbow", "Beast-Claw Gauntlet", "Dual Throwing Axes"],
    "Warlock": ["Cursed Scythe", "Demon-Skull Catalyst", "Void-Touched Grimoire"]
}

def buat_char():
    nama_depan_2 = random.choice(pilihan_nama_depan_2)
    nama_belakang_2 = random.choice(pilihan_nama_belakang_2)
    kelas = random.choice(pilihan_kelas_2)

    nama_lengkap_2 = nama_depan_2 + " " + nama_belakang_2
    senjata_tersedia_2 = pilihan_senjata_2[kelas]
    senjata_pilihan_2 = random.choice(senjata_tersedia_2)

    point_total = 100
    power = random.randint(1, 80)
    point_left = point_total - power
    magic = random.randint(1, point_left)
    agility = point_left - magic

    if kelas == "Warrior":
        power += 10
    elif kelas == "Mage":
        magic += 10
    elif kelas == "Rogue":
        agility += 10
    elif kelas == "Healer":
        magic += 5
        power += 5
    elif kelas == "Paladin":
        power += 8
        magic += 8
    elif kelas == "Hunter":
        agility += 10
        power += 5
    elif kelas == "Warlock":
        magic += 15

    print("Nama:", nama_lengkap_2)
    print("Kelas:", kelas)
    print("Senjata:", senjata_pilihan_2)
    print(f"Stats -> Power: {power}, Magic: {magic}, Agility: {agility}")

print("--- Menciptakan Karakter Uji Coba #1 ---")
buat_char()

print("\n--- Menciptakan Karakter Uji Coba #2 ---")
buat_char()

print("\n--- Selamat Berpetualang ---")
```
