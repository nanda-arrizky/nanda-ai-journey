```python
import pandas as pd
import numpy as np
# Data Mentah Pelanggan
data_pelanggan = {
    'ID_Pelanggan': [101, 102, 103, 104, 105, 106],
    'Nama': ['Adi', 'Budi', 'Cici', 'Dedi', 'Eka', 'Fani'],
    'Kota': ['Jakarta', 'Bandung', 'Jakarta', 'Surabaya', 'Bandung', 'Yogyakarta']
}
df_pelanggan = pd.DataFrame(data_pelanggan)
df_pelanggan
```

```python
display(df_pelanggan.head())
df_pelanggan.info()
df_pelanggan.describe()
```

```python
# Data Mentah Transaksi
data_transaksi = {
    'ID_Transaksi': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    'ID_Pelanggan': [101, 103, 102, 105, 101, 104, 107],
    'Jumlah_Belanja': [500000, 750000, 120000, np.nan, 250000, 300000, 450000]
}
df_transaksi = pd.DataFrame(data_transaksi)
df_transaksi
```

```python
display(df_transaksi.head())
df_transaksi.info()
df_transaksi.describe()
```

```python
df_transaksi_bersih = df_transaksi.copy()
rerata_jumlah_belanja = df_transaksi['Jumlah_Belanja'].mean()
df_transaksi_bersih['Jumlah_Belanja'] = df_transaksi_bersih['Jumlah_Belanja'].fillna(rerata_jumlah_belanja)
df_transaksi_bersih.info()
```

```python
df_gabungan = pd.merge(df_pelanggan, df_transaksi_bersih, on='ID_Pelanggan', how='inner')
df_gabungan
```

```python
laporan_per_kota = df_gabungan.groupby('Kota').agg(
    Total_Belanja_Kota=('Jumlah_Belanja', 'sum')
)
laporan_per_kota_terurut = laporan_per_kota.sort_values(by='Total_Belanja_Kota', ascending=False)
display(laporan_per_kota_terurut)
peringkat_satu = laporan_per_kota_terurut.iloc[0]
print(f"\nJadi, kota dengan total belanja tertinggi adalah {peringkat_satu.name} dengan nilai: Rp{peringkat_satu.iloc[0]:,.0f}")
```
