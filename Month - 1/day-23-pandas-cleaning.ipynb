```python
import pandas as pd
import numpy as np
```

```python
data_inventaris = {
    'Nama_Item': ['Pedang Cahaya', 'Perisai Abadi', 'Jubah Bayangan', 'Tongkat Petir', 'Cincin Kehidupan'],
    'Tipe': ['Senjata', 'Perisai', 'Armor', 'Senjata', 'Aksesoris'],
    'Kekuatan': [120, 0, 15, 150, 5],
    'Harga_Jual': [1500.0, 2200.0, 1800.0, 1750.5, 3000.0],
    'Jumlah_Stok': [15, 8, 12, 10, 30]
}
df_items = pd.DataFrame(data_inventaris)

print("--- DataFrame Awal ---")
display(df_items)
```

```python
df_items['Rasio_Kekuatan_Harga'] = df_items['Kekuatan'] / df_items['Harga_Jual']
print("\n--- DataFrame Setelah Menambah Kolom 'Rasio' ---")
display(df_items) #Karena kita menambah data, jadi kita perlu menampilkan semuanya agar bisa meihat perubahannya

df_items_dihapus = df_items.drop(columns=['Jumlah_Stok'])
print("\n--- DataFrame Setelah Menghapus Kolom 'Jumlah_Stok' ---")
display(df_items_dihapus) #karena cuma untuk contoh, jadi kita masih perlu datanya. Oleh karena itu kita membedakan dataframenya dengan variabel dulu
```

```python
data_inventaris_kotor = {
    'Nama_Item': ['Pedang Cahaya', 'Perisai Abadi', 'Jubah Bayangan', 'Tongkat Petir', 'Cincin Kehidupan'],
    'Tipe': ['Senjata', 'Perisai', 'Armor', 'Senjata', np.nan], # Tipe Cincin hilang
    'Kekuatan': [120, 0, 15, 150, 5],
    'Harga_Jual': [1500.0, np.nan, 1800.0, 1750.5, 3000.0], # Harga Perisai hilang
    'Jumlah_Stok': [15, 8, 12, np.nan, 30] # Stok Tongkat Petir hilang
}
df_kotor = pd.DataFrame(data_inventaris_kotor)

print("--- DataFrame 'Kotor' dengan Data Hilang ---")
display(df_kotor)
```

```python
print("\n--- Peta Data Hilang (.isnul()) ---")
display(df_kotor.isnull()) #Kita bertanya, apakah ada yang nilainya null(NaN)?

print("\n--- Jumlah Data Hilang per Kolom ---")
print(df_kotor.isnull().sum())
```

```python
df_dibuang = df_kotor.dropna()
print("\n--- Hasil .dropna() (Membuang Baris) ---")
display(df_dibuang)

df_diisi = df_kotor.fillna(0)
print("\n--- Hasil .fillna(0) (Mengisi dengan Nol) ---")
display(df_diisi)
```

```python
df_tanpa_perisai = df_kotor.drop(labels=[1])
print("--- df_kotor setelah menghapus baris indeks 1 ---")
display(df_tanpa_perisai)

df_items_indexed = df_items.set_index('Nama_Item')
df_tanpa_tongkat = df_items_indexed.drop(labels=['Tongkat Petir'])
print("\n--- DataFrame setelah menghapus 'Tongkat Petir' ---")
display(df_tanpa_tongkat)
```

```python
kondisi_tipe = (df_items['Tipe'] == 'Senjata')
kondisi_kekuatan = (df_items['Kekuatan'] > 130)

senjata_kuat = df_items[kondisi_tipe & kondisi_kekuatan]
print("--- Senjata dengan Kekuatan > 130 ---")
display(senjata_kuat)

kondisi_perisai = (df_items['Tipe'] == 'Perisai')
kondisi_armor = (df_items['Tipe'] == 'Armor')

item_defensif = df_items[kondisi_perisai | kondisi_armor]
print("\n--- Semua Item Defensif (Perisai atau Armor) ---")
display(item_defensif)
```

```python
# Versi inline untuk filter DAN
senjata_kuat_langsung = df_items[(df_items['Tipe'] == 'Senjata') & (df_items['Kekuatan'] > 130)]

print("--- Senjata Kuat (Versi Inline) ---")
display(senjata_kuat_langsung)


# Versi inline untuk filter ATAU
item_defensif_langsung = df_items[(df_items['Tipe'] == 'Perisai') | (df_items['Tipe'] == 'Armor')]

print("\n--- Item Defensif (Versi Inline) ---")
display(item_defensif_langsung)
```

```python
data_karakter = {
    'Nama': ['Arion', 'Lyra', 'Gideon', 'Seraphina', 'Kael'],
    'Kelas': ['Warrior', 'Mage', 'Guardian', 'Ranger', 'Guardian'],
    'Attack': [150, 180, 80, 120, 95],
    'Defense': [180, 120, 250, 150, 220],
    'HP': [2000, 1500, 3000, 1800, 2800]
}

df_karakter = pd.DataFrame(data_karakter)
df_karakter
```

```python
df_karakter_indexed = df_karakter.set_index('Nama')
kandidat_tank = df_karakter_indexed[(df_karakter_indexed['Defense'] > 200) & (df_karakter_indexed['Attack'] < 100)]
display(kandidat_tank.loc[:, ['HP']])
#Aku tambahkan untuk yang paling cocok jadi tank
#Parameternya yang aku gunakan adalah nilai def tertinggi
def_tertinggi = kandidat_tank['Defense'].max()
tank_pilihan = kandidat_tank['Defense'] == def_tertinggi
display(kandidat_tank.loc[tank_pilihan])
```
