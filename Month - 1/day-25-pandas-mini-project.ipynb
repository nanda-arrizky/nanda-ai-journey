```python
import pandas as pd
import numpy as np

data_penjualan = {
    'Nama_Game': ['Starlight Odyssey', 'Cyberforge', 'Kingdoms of Ethera', 'Wasteland Drifters', 'Abyssal Echoes', 'Celestial Era'],
    'Genre': ['RPG', 'Action', 'RPG', 'RPG', 'Action', 'Strategy'],
    'Tahun_Rilis': [2022, 2023, 2022, 2021, np.nan, 2023],
    'Harga_Satuan': [60, 55, 60, 45, 55, 50],
    'Unit_Terjual': [500000, 850000, 720000, 450000, 950000, np.nan]
}

df_penjualan = pd.DataFrame(data_penjualan)
```

```python
display(df_penjualan.head())
display(df_penjualan.info())
display(df_penjualan.describe())
```

```python
df_bersih = df_penjualan.copy() #Seperti biasa kita copy dulu
median_tahun = df_bersih['Tahun_Rilis'].median()
mean_unit = df_bersih['Unit_Terjual'].mean()
print(f"Median Tahun Rilis: {median_tahun}")
print(f"Rerata Unit Terjual: {mean_unit}")

df_bersih['Tahun_Rilis'] = df_bersih['Tahun_Rilis'].fillna(median_tahun)
df_bersih['Unit_Terjual'] = df_bersih['Unit_Terjual'].fillna(mean_unit)
print("\n--- Info DataFrame Setelah Pembersihan ---")
df_bersih.info()
print("\n--- DataFrame Bersih ---")
display(df_bersih)
```

```python
df_bersih['Pendapatan_Total'] = df_bersih['Harga_Satuan'] * df_bersih['Unit_Terjual']
print("--- DataFrame dengan kolom Baru 'Pendapatan_Total'")
display(df_bersih)
```

```python
laporan_per_genre = df_bersih.groupby('Genre').agg(
    Total_Pendapatan=('Pendapatan_Total', 'sum'),
    Rerata_Unit_Terjual=('Unit_Terjual', 'mean')
)
laporan_final_terurut = laporan_per_genre.sort_values(by='Total_Pendapatan', ascending=False)

print("--- Laporan Final per Genre (diurutkan) ---")
display(laporan_final_terurut)
```
